• The program should:

• Conversion Options:
1. Length:
– Convert meters (m) to feet (ft).
– Convert feet (ft) to meters (m).
2. Weight:
– Convert kilograms (kg) to pounds (lbs).
– Convert pounds (lbs) to kilograms (kg).
3. Volume:
– Convert liters (L) to gallons (gal).
– Convert gallons (gal) to liters (L).

```python
def meters_to_feet(meters):
    """
    Convert meters to feet.

    Parameters:
    meters (float): The length in meters.

    Returns:
    float: The length in feet.
    """
    return meters * 3.28084

def feet_to_meters(feet):
    return feet / 3.28084

def kilograms_to_pounds(kilograms):
    return kilograms * 2.20462

def pounds_to_kilograms(pounds):
    return pounds / 2.20462

def liters_to_gallons(liters):
    return liters * 0.264172

def gallons_to_liters(gallons):
    return gallons / 0.264172

def get_conversion_choice():
    print("Choose the type of conversion:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")
    choice = input("Enter your choice: ")
    return choice

def get_length_conversion_choice():
    print("Choose the length conversion:")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    choice = input("Enter your choice: ")
    return choice

def get_weight_conversion_choice():
    print("Choose the weight conversion:")
    print("1. Kilograms to Pounds")
    print("2. Pounds to Kilograms")
    choice = input("Enter your choice: ")
    return choice

def get_volume_conversion_choice():
    print("Choose the volume conversion:")
    print("1. Liters to Gallons")
    print("2. Gallons to Liters")
    choice = input("Enter your choice: ")
    return choice

def main():
    try:
        conversion_type = get_conversion_choice()

        if conversion_type == '1':
            length_choice = get_length_conversion_choice()
            value = float(input("Enter the length value to be converted: "))
            if length_choice == '1':
                result = meters_to_feet(value)
                print(f"{value} meters is equal to {result:.2f} feet.")
            elif length_choice == '2':
                result = feet_to_meters(value)
                print(f"{value} feet is equal to {result:.2f} meters.")
            else:
                print("Invalid length conversion choice.")

        elif conversion_type == '2':
            weight_choice = get_weight_conversion_choice()
            value = float(input("Enter the weight value to be converted: "))
            if weight_choice == '1':
                result = kilograms_to_pounds(value)
                print(f"{value} kilograms is equal to {result:.2f} pounds.")
            elif weight_choice == '2':
                result = pounds_to_kilograms(value)
                print(f"{value} pounds is equal to {result:.2f} kilograms.")
            else:
                print("Invalid weight conversion choice.")

        elif conversion_type == '3':
            volume_choice = get_volume_conversion_choice()
            value = float(input("Enter the volume value to be converted: "))
            if volume_choice == '1':
                result = liters_to_gallons(value)
                print(f"{value} liters is equal to {result:.2f} gallons.")
            elif volume_choice == '2':
                result = gallons_to_liters(value)
                print(f"{value} gallons is equal to {result:.2f} liters.")
            else:
                print("Invalid volume conversion choice.")

        else:
            print("Invalid conversion type choice.")

    except ValueError:
        print("Invalid input. Please enter a numeric value.")

if __name__ == "__main__":
    main()
```

Create a Python program that performs various mathematical operations on a list of numbers.

```python
def calculate_sum(numbers):
    """
    Calculate the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)

def calculate_average(numbers):
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    return max(numbers)

def find_minimum(numbers):
    return min(numbers)

def get_numbers_from_user():
    input_string = input("Enter numbers separated by spaces: ")
    number_strings = input_string.split()
    numbers = [float(num) for num in number_strings]
    return numbers

def main():
    try:
        print("Choose an operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")
        choice = input("Enter the number corresponding to your choice: ")

        numbers = get_numbers_from_user()
        if not numbers:
            raise ValueError("The list of numbers is empty.")

        if choice == '1':
            result = calculate_sum(numbers)
            print(f"The sum of the numbers is: {result}")
        elif choice == '2':
            result = calculate_average(numbers)
            print(f"The average of the numbers is: {result}")
        elif choice == '3':
            result = find_maximum(numbers)
            print(f"The maximum value is: {result}")
        elif choice == '4':
            result = find_minimum(numbers)
            print(f"The minimum value is: {result}")
        else:
            print("Invalid choice. Please select a valid operation.")

    except ValueError as e:
        print(f"Invalid input: {e}")
    except ZeroDivisionError:
        print("Cannot calculate the average of an empty list.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```

# **4.2 Exercise on List Manipulation:**

```python
def extract_every_other(lst):
    """Return a new list containing every other element from the original list."""
    return lst[::2]

def get_sublist(lst, start, end):
    """Return a sublist from the original list, from index 'start' to 'end' inclusive."""
    return lst[start:end+1]

def reverse_list(lst):
    """Return a new list that is the reverse of the original list."""
    return lst[::-1]

def remove_first_last(lst):
    """Return a new list with the first and last elements removed from the original list."""
    return lst[1:-1]

def get_first_n(lst, n):
    """Return a new list containing the first 'n' elements of the original list."""
    return lst[:n]

def get_last_n(lst, n):
    """Return a new list containing the last 'n' elements of the original list."""
    return lst[-n:]

def reverse_skip(lst):
    """Return a new list containing every second element from the original list, starting from the second-to-last, moving backward."""
    return lst[-2::-2]

def main():
    options = {
        '1': ('Extract every other element', extract_every_other),
        '2': ('Slice a sublist', get_sublist),
        '3': ('Reverse the list', reverse_list),
        '4': ('Remove the first and last elements', remove_first_last),
        '5': ('Get the first n elements', get_first_n),
        '6': ('Get the last n elements', get_last_n),
        '7': ('Extract elements in reverse order skipping one', reverse_skip)
    }

    while True:
        print("\nChoose a list manipulation operation:")
        for key, (description, _) in options.items():
            print(f"{key}. {description}")
        print("8. Exit")

        choice = input("Enter the number of your choice: ")

        if choice == '8':
            print("Exiting the program.")
            break

        if choice not in options:
            print("Invalid choice. Please try again.")
            continue

        try:
            lst = input("Enter a list of elements separated by spaces: ").split()
            lst = [int(x) if x.isdigit() else x for x in lst]  # Convert numeric strings to integers

            if choice in ['2', '5', '6']:
                if choice == '2':
                    start = int(input("Enter the starting index: "))
                    end = int(input("Enter the ending index: "))
                    result = options[choice][1](lst, start, end)
                else:
                    n = int(input("Enter the number of elements (n): "))
                    result = options[choice][1](lst, n)
            else:
                result = options[choice][1](lst)

            print(f"Result: {result}")

        except ValueError as e:
            print(f"Invalid input: {e}")
        except IndexError as e:
            print(f"Index error: {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```

# **4.3 Exercise on Nested List:**

```python
def flatten(lst):
    """Flattens a nested list into a single-dimensional list."""
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list

def access_nested_element(lst, indices):
    """Accesses a specific element in a nested list given a list of indices."""
    element = lst
    for index in indices:
        element = element[index]
    return element

def sum_nested(lst):
    """Calculates the sum of all numbers in a nested list, regardless of depth."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

def remove_element(lst, elem):
    """Removes all occurrences of a specific element from a nested list."""
    if isinstance(lst, list):
        return [remove_element(item, elem) for item in lst if item != elem]
    else:
        return lst

def find_max(lst):
    """Finds the maximum element in a nested list, regardless of depth."""
    max_elem = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_elem = max(max_elem, find_max(item))
        else:
            max_elem = max(max_elem, item)
    return max_elem

def count_occurrences(lst, elem):
    """Counts how many times a specific element appears in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

def deep_flatten(lst):
    """Flattens a list of lists of lists into a single list, regardless of depth."""
    return flatten(lst)

def average_nested(lst):
    """Calculates the average of all elements in a nested list."""
    flat_list = flatten(lst)
    return sum(flat_list) / len(flat_list) if flat_list else 0

if __name__ == "__main__":
    nested_list = [[1, 2], [3, [4, 5]], 6]
    print("Flattened list:", flatten(nested_list))
    print("Accessed element [1, 1]:", access_nested_element(nested_list, [1, 1]))
    print("Sum of elements:", sum_nested(nested_list))
    print("List after removing 2:", remove_element(nested_list, 2))
    print("Maximum element:", find_max(nested_list))
    print("Occurrences of 2:", count_occurrences(nested_list, 2))
    print("Deep flattened list:", deep_flatten(nested_list))
    print("Average of elements:", average_nested(nested_list))
```

Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” func-
tion of Numpy}

```python
import numpy as np

# Problem 4: Matrix Operations
print("Problem 4: Matrix Operations")

# Defining matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# 1. Prove A.A^(-1) = I
print("\n1. Proving A.A^(-1) = I")
A_inv = np.linalg.inv(A)
print("A^(-1):\n", A_inv)
print("A.A^(-1):\n", np.dot(A, A_inv))
print("This is approximately the identity matrix I (small differences due to floating-point precision)")

# 2. Prove AB ≠ BA
print("\n2. Proving AB ≠ BA")
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB:\n", AB)
print("BA:\n", BA)
print("AB = BA:", np.array_equal(AB, BA))
print("Therefore, AB ≠ BA is proven")

# 3. Prove (AB)^T = B^T.A^T
print("\n3. Proving (AB)^T = B^T.A^T")
AB_transpose = np.transpose(np.dot(A, B))
print("(AB)^T:\n", AB_transpose)

B_transpose = np.transpose(B)
A_transpose = np.transpose(A)
BT_AT = np.dot(B_transpose, A_transpose)
print("B^T.A^T:\n", BT_AT)
print("(AB)^T = B^T.A^T:", np.array_equal(AB_transpose, BT_AT))
print("Therefore, (AB)^T = B^T.A^T is proven")

# Solving the system of linear equations
print("\n------ Solving the system of linear equations ------")
print("2x - 3y + z = -1")
print("x - y + 2z = -3")
print("3x + y - z = 9")

# Setting up the coefficient matrix A and constants vector B
coefficients = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

constants = np.array([-1, -3, 9])

print("\nMatrix form: AX = B")
print("Coefficient matrix A:\n", coefficients)
print("Constants vector B:", constants)

# Solving using inverse method manually
print("\nSolving using inverse method manually:")
coef_inv = np.linalg.inv(coefficients)
print("A^(-1):\n", coef_inv)
solution_manual = np.dot(coef_inv, constants)
print("Solution X = A^(-1)B:", solution_manual)

# Solving using np.linalg.solve function
print("\nSolving using np.linalg.solve function:")
solution_solve = np.linalg.solve(coefficients, constants)
print("Solution X:", solution_solve)

# Verifying the solution
print("\nVerifying the solution:")
print("AX =", np.dot(coefficients, solution_solve))
print("This should match B:", constants)
```

10.2 Experiment: How Fast is Numpy?
In this exercise, you will compare the performance and implementation of operations using plain Python
lists (arrays) and NumPy arrays. Follow the instructions:
1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for

2. Element-wise Multiplication
• Using Python Lists, perform element-wise multiplication of two lists of size 1, 000, 000.
Measure and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.
3. Dot Product
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for

4. Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
import numpy as np
import time
import random

print("Speed Comparison: Python Lists vs NumPy Arrays")
print("="*50)

# Function to measure execution time
def time_execution(func, *args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    return result, end_time - start_time

# 1. Element-wise Addition
print("\n1. Element-wise Addition (size: 1,000,000)")
print("-"*50)

# Create lists and arrays of size 1,000,000
size = 1000000
list1 = [random.random() for _ in range(size)]
list2 = [random.random() for _ in range(size)]
array1 = np.array(list1)
array2 = np.array(list2)

# Python list element-wise addition
def list_addition(l1, l2):
    return [a + b for a, b in zip(l1, l2)]

_, list_add_time = time_execution(list_addition, list1, list2)
print(f"Python List Addition Time: {list_add_time:.6f} seconds")

# NumPy array element-wise addition
_, numpy_add_time = time_execution(np.add, array1, array2)
print(f"NumPy Array Addition Time: {numpy_add_time:.6f} seconds")
print(f"NumPy is {list_add_time/numpy_add_time:.2f}x faster")

# 2. Element-wise Multiplication
print("\n2. Element-wise Multiplication (size: 1,000,000)")
print("-"*50)

# Python list element-wise multiplication
def list_multiplication(l1, l2):
    return [a * b for a, b in zip(l1, l2)]

_, list_mul_time = time_execution(list_multiplication, list1, list2)
print(f"Python List Multiplication Time: {list_mul_time:.6f} seconds")

# NumPy array element-wise multiplication
_, numpy_mul_time = time_execution(np.multiply, array1, array2)
print(f"NumPy Array Multiplication Time: {numpy_mul_time:.6f} seconds")
print(f"NumPy is {list_mul_time/numpy_mul_time:.2f}x faster")

# 3. Dot Product
print("\n3. Dot Product (size: 1,000,000)")
print("-"*50)

# Python list dot product
def list_dot_product(l1, l2):
    return sum(a * b for a, b in zip(l1, l2))

_, list_dot_time = time_execution(list_dot_product, list1, list2)
print(f"Python List Dot Product Time: {list_dot_time:.6f} seconds")

# NumPy array dot product
_, numpy_dot_time = time_execution(np.dot, array1, array2)
print(f"NumPy Array Dot Product Time: {numpy_dot_time:.6f} seconds")
print(f"NumPy is {list_dot_time/numpy_dot_time:.2f}x faster")

# 4. Matrix Multiplication
print("\n4. Matrix Multiplication (size: 1000x1000)")
print("-"*50)

# Creating smaller matrices for practical reasons
matrix_size = 100  # Using 100x100 instead of 1000x1000 to make it run in reasonable time
print(f"Note: Using {matrix_size}x{matrix_size} matrices instead of 1000x1000 for practical demonstration")

# Create Python lists as matrices
list_matrix1 = [[random.random() for _ in range(matrix_size)] for _ in range(matrix_size)]
list_matrix2 = [[random.random() for _ in range(matrix_size)] for _ in range(matrix_size)]

# Create NumPy array matrices
numpy_matrix1 = np.array(list_matrix1)
numpy_matrix2 = np.array(list_matrix2)

# Python list matrix multiplication
def list_matrix_multiply(m1, m2):
    result = [[0 for _ in range(len(m2[0]))] for _ in range(len(m1))]
    for i in range(len(m1)):
        for j in range(len(m2[0])):
            for k in range(len(m2)):
                result[i][j] += m1[i][k] * m2[k][j]
    return result

_, list_matmul_time = time_execution(list_matrix_multiply, list_matrix1, list_matrix2)
print(f"Python List Matrix Multiplication Time: {list_matmul_time:.6f} seconds")

# NumPy array matrix multiplication
_, numpy_matmul_time = time_execution(np.matmul, numpy_matrix1, numpy_matrix2)
print(f"NumPy Array Matrix Multiplication Time: {numpy_matmul_time:.6f} seconds")
print(f"NumPy is {list_matmul_time/numpy_matmul_time:.2f}x faster")

# Summary
print("\nSUMMARY")
print("="*50)
print(f"Element-wise Addition: NumPy is {list_add_time/numpy_add_time:.2f}x faster")
print(f"Element-wise Multiplication: NumPy is {list_mul_time/numpy_mul_time:.2f}x faster")
print(f"Dot Product: NumPy is {list_dot_time/numpy_dot_time:.2f}x faster")
print(f"Matrix Multiplication: NumPy is {list_matmul_time/numpy_matmul_time:.2f}x faster")
print("\nNote: The exact speed ratios may vary based on your hardware and system load.")
print("="*50)
```
